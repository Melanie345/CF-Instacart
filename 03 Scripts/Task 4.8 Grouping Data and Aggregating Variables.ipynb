# 06 Create a spending flag for each user based on the average price across all their orders using the specialized criteria listed at the beginning of this notebook

```python
#create a column for an average spender
ords_prods_merge['average_spend'] = ords_prods_merge.groupby(['user_id'])['prices'].transform("mean")
```

```python
#showing the newly created average spend column
ords_prods_merge.head()
```

```python
#creating the flags for high and low spenders, first the low spender
ords_prods_merge.loc[ords_prods_merge['average_spend'] <10, 'spender_flag'] = 'low_spender'
```

```python
ords_prods_merge.loc[ords_prods_merge['average_spend'] >=10, 'spender_flag'] = 'high_spender'
```

```python
#let's look at the newly made spender flags
ords_prods_merge['spender_flag'].value_counts(dropna = False)
```

```python
#lets only look at the specific columns of user_id, average spend, spender_flag and prices 
ords_prods_merge[['user_id', 'average_spend', 'spender_flag', 'prices']].head(10)
```

# 07 Create an order frequency flag

```python
ords_prods_merge['customer_frequency'] = ords_prods_merge.groupby(['user_id'])['days_since_prior_order'].transform("median")
```

```python
#lets show the new customer frequency column
ords_prods_merge.head()
```

```python
#creating flag for order frequency, first non frequent, then the regular followed by frequent customer
ords_prods_merge.loc[ords_prods_merge['customer_frequency'] >20, 'frequency_flag'] = 'Non-frequent customer'
```

```python
ords_prods_merge.loc[(ords_prods_merge['customer_frequency'] >10) & (ords_prods_merge['customer_frequency'] <=20),'frequency_flag'] = 'Regular customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['customer_frequency'] <=10, 'frequency_flag'] = 'Frequent customer'
```

```python
#lets look at the newly made frequency flags
ords_prods_merge['frequency_flag'].value_counts(dropna = False)
```

```python
#checking the specific columns of user id, customer frequency, frequency flag
ords_prods_merge[['user_id', 'customer_frequency', 'frequency_flag']].head(60)
```
