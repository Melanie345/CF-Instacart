## Step 1 Create Price_label and Busiest_day columns

```python
#creating subset of first 1,000,000 rows
df = df_ords_prods_merge[:1000000]
```

```python
df.shape
```

```python
#Define a function for price_label column
def price_label(row):
    
    if row['prices'] <= 5:
        return 'Low-range product'
    elif (row['prices'] > 5) and (row['prices'] <= 15):
        return 'Mid-range product'
    elif row['prices'] > 15:
        return 'High range'
    else: return 'Not enough data'
```

```python
#apply new price_range function, this wants me to use .loc 
df['price_range'] = df.apply(price_label, axis=1)
```

```python
df['price_range'].value_counts(dropna = False)
```

```python
df['prices'].max()
```

```python
#use loc() method to use price_range function, first, high-range product
df.loc[df['prices'] > 15, 'price_range_loc'] = 'High-range product'
```

```python
#next the mid-range
df.loc[(df['prices'] <= 15) & (df['prices'] > 5), 'price_range_loc'] = 'Mid-range product' 
```

```python
#finally, the low-range...
df.loc[df['prices'] <= 5, 'price_range_loc'] = 'Low-range product'
```

```python
df['price_range_loc'].value_counts(dropna = False)
```

```python
#using the loc process on the entire df_ords_prods_merge
df_ords_prods_merge.loc[df_ords_prods_merge['prices'] > 15, 'price_range_loc'] = 'High-range product'
```

```python
#next the#mid-range
df_ords_prods_merge.loc[(df_ords_prods_merge['prices'] <= 15) & (df_ords_prods_merge['prices'] > 5), 'price_range_loc'] = 'Mid-range product' 
```

```python
#finding the low range on the entire df_ords_prods_merge set
df_ords_prods_merge.loc[df_ords_prods_merge['prices'] <= 5, 'price_range_loc'] = 'Low-range product'
```

```python
df_ords_prods_merge['price_range_loc'].value_counts(dropna = False)
```

### Step 1b Create busiest_day column

```python
#find frequency of orders_day_of_week
df_ords_prods_merge['orders_day_of_week'].value_counts(dropna = False)
```

```python
#creating a new values for busiest_day column that will show busiest day, least busy or regularly busy
result = []

for value in df_ords_prods_merge["orders_day_of_week"]:
  if value == 0:
    result.append("Busiest day")
  elif value == 4:
    result.append("Least busy")
  else:
    result.append("Regularly busy")
```

```python
result
```

```python
#creating the 'busiest_day' column in df_ords_prods_merge
df_ords_prods_merge['busiest_day'] = result
```

```python
df_ords_prods_merge['busiest_day'].value_counts(dropna = False)
```

#Step 4 Create new columnn 'busiest_period_of_day" to identify periods of time "most orders", "average orders", "fewest orders"

```python
#check value counts in orders_time_of_day']
df_ords_prods_merge['orders_time_of_day'].value_counts()
```

```python
#create for-loop if statement labeling periods of time "most orders", "average orders", "fewest orders"

result_2 = []

for value in df_ords_prods_merge["orders_time_of_day"]:
  if value in [10, 11, 14, 15, 13, 12, 16, 9]:
    result_2.append("Most orders")
  elif value in [23, 6, 0, 1, 5, 2, 4, 3]:
    result_2.append("Fewest orders")
  else:
    result_2.append("Average orders")
```

```python
result_2
```

```python
#create new column of busiest_period_of_day

df_ords_prods_merge['busiest_period_of_day'] = result_2
```
