#Step 5 Combine products data set with orders products combined set

```python
#dropping un-named column in df_prods
df_prods = df_prods.drop(columns = ['Unnamed: 0'])
```

```python
#evidence of un-named column dropped
df_prods.head()
```

```python
#dropping un-named columns in the df_orders_products_combined and also the eval_set that somehow snuck back in... 
df_orders_products_combined = df_orders_products_combined.drop(columns = ['Unnamed: 0.1', 'Unnamed: 0', 'eval_set'])
```

```python
#evidence of unnamed column & eval set dropped
df_orders_products_combined.head()
```

```python
#Merging df_prods and df_orders_products_combined!
df_merged = df_prods.merge(df_orders_products_combined, on = 'product_id')
```
